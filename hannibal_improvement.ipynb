```python
import pandas as pd
```

```python
#open files
hannibal =  pd.read_csv('hannibal_df.csv', sep=',')
```

```python
#converting NaN values to 0 and changing floats to int and converting postdate to date time format
hannibal['Likes'] = hannibal['Likes'].fillna(0)
hannibal['Likes'] = hannibal['Likes'].astype(int)
hannibal['Reviews'] = hannibal['Reviews'].fillna(0)
hannibal['Reviews'] = hannibal['Reviews'].astype(int)
hannibal['Postdate'] = pd.to_datetime(hannibal['Postdate'])
```

**Sum of likes by year**

```python
hannibal['Likes'].groupby([hannibal['Postdate'].dt.year]).sum().plot.bar()
```

**Sum of collection saves**

```python
hannibal['Collections'].groupby([hannibal['Postdate'].dt.year]).sum().plot.bar()
```

**Sum of story followers**

```python
hannibal['Followers'].groupby([hannibal['Postdate'].dt.year]).sum().plot.bar()
```

```python
hannibal['Reviews'].groupby([hannibal['Postdate'].dt.year]).sum().plot.bar()
```
