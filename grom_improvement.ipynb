```python
import pandas as pd
```

```python
grom =  pd.read_csv('grom_df.csv', sep=',')
```

```python
#converting NaN values to 0 and changing floats to int and converting postdate to date time format
grom['Likes'] = grom['Likes'].fillna(0)
grom['Likes'] = grom['Likes'].astype(int)
grom['Reviews'] = grom['Reviews'].fillna(0)
grom['Reviews'] = grom['Reviews'].astype(int)
grom['Postdate'] = pd.to_datetime(grom['Postdate'])
```

**Sum of likes by months in years 2021 and 2022**

```python
grom_2021 = grom[grom['Postdate'].dt.year.isin([2021])]
grom_2021 = grom_2021.reset_index(drop=True)
grom_2022 = grom[grom['Postdate'].dt.year.isin([2022])]
grom_2022 = grom_2022.reset_index(drop=True)
```

**2021**

```python
grom_2021['Likes'].groupby([grom_2021['Postdate'].dt.month]).sum().plot.bar()
```

**2022**

```python
grom_2022['Likes'].groupby([grom_2022['Postdate'].dt.month]).sum().plot.bar()
```

**Collections 2021**

```python
grom_2021.groupby([grom_2021['Postdate'].dt.month])['Collections'].sum().plot.bar()
```

**Collections 2022**

```python
grom_2022.groupby([grom_2022['Postdate'].dt.month])['Collections'].sum().plot.bar()
```
