```python
import pandas as pd
```

```python
witcher_books_series =  pd.read_csv('witcher_books_series.csv', sep=',')
```

```python
witcher_games1 = pd.read_csv('witcher_games.csv', sep=',')
witcher_games2 = pd.read_csv('witcher_games_2.csv', sep=',')
witcher_games3 = pd.read_csv('witcher_games_3.csv', sep=',')
witcher_games4 = pd.read_csv('witcher_games_4.csv', sep=',')
```

```python
witcher_games = pd.concat([witcher_games1, witcher_games2, witcher_games3, witcher_games4],ignore_index=True)
```

```python
witcher_df = pd.concat([witcher_books_series, witcher_games],ignore_index=True)
```

```python
witcher_df = witcher_df.drop_duplicates()
```

```python
witcher_df = witcher_df.reset_index(drop=True)
```

```python
witcher_df.to_csv("witcher_df.csv", sep=',', index=False)
```

```python
#converting NaN values to 0 and changing floats to int and converting postdate to date time format
witcher_df['Likes'] = witcher_df['Likes'].fillna(0)
witcher_df['Likes'] = witcher_df['Likes'].astype(int)
witcher_df['Reviews'] = witcher_df['Reviews'].fillna(0)
witcher_df['Reviews'] = witcher_df['Reviews'].astype(int)
witcher_df['Postdate'] = pd.to_datetime(witcher_df['Postdate'])
```

**Sum of likes by year**

```python

witcher_df['Likes'].groupby([witcher_df['Postdate'].dt.year]).sum().plot.bar()
```

**Sum of reviews by year**

```python
witcher_df['Reviews'].groupby([witcher_df['Postdate'].dt.year]).sum().plot.bar()
```

**collections**

```python
witcher_df['Collections'].groupby([witcher_df['Postdate'].dt.year]).sum().plot.bar()
```

```python
witcher_df['Followers'].groupby([witcher_df['Postdate'].dt.year]).sum().plot.bar()
```
