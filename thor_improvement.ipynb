```python
import pandas as pd
```

```python
thor_df =  pd.read_csv('thor_df.csv', sep=',')
```

```python
thor_df['Postdate'] = pd.to_datetime(thor_df['Postdate'])
```

**Sum of likes by year**

```python
thor_df['Likes'].groupby([thor_df['Postdate'].dt.year]).sum().plot.bar()
```

**Sum of reviews by year**

```python
thor_df['Reviews'].groupby([thor_df['Postdate'].dt.year]).sum().plot.bar()
```

**collections**

```python
thor_df['Collections'].groupby([thor_df['Postdate'].dt.year]).sum().plot.bar()
```

**Followers**

```python
thor_df['Followers'].groupby([thor_df['Postdate'].dt.year]).sum().plot.bar()
```
